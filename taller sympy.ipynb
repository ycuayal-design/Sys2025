```python
import sympy as sym
t, omega = sym.symbols('t omega', real=True)

# Definimos una señal sinusoidal como ejemplo
f_signal = sym.sin(omega * t)

# Aplicamos el valor absoluto de la señal para rectificarla (rectificador de onda completa)
rectified_signal = sym.Abs(f_signal)

# Graficamos la señal original y la rectificada
p1 = sym.plot(f_signal.subs(omega, 2), (t, 0, 4 * sym.pi), show=False, ylabel='Señal original')
p2 = sym.plot(rectified_signal.subs(omega, 2), (t, 0, 4 * sym.pi), show=False, ylabel='Señal rectificada', line_color='red')

p1.extend(p2)
p1.show()
```

```python
tau = 5
# Define the signal x
x = f_signal.subs(omega, 2) # Using the signal from the previous cell

y_shifted = x.subs(t, tau - t)

# Plotting the original and shifted signals
px = sym.plot(x, (t, -2, 8), ylim=[-0.2, 1.2], title='Original and Shifted Signal', ylabel=r'$y(t)$', line_color='gray', show=False)
py = sym.plot(y_shifted, (t, -2, 8), ylim=[-0.2, 1.2], ylabel=r'$y(t)$', show=False)
px.extend(py)
px.show()
```
